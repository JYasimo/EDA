```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score  # Import the f1_score function
from sklearn.linear_model import LogisticRegression
from skopt import BayesSearchCV
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler

# Define the number of folds for cross-validation
n_folds = 5

# Initialize lists to store the estimated parameters and validation scores from each fold
params = []
params_intervals = []
train_f1_scores = []  # Change from train_auroc_scores to train_f1_scores
test_f1_scores = []   # Change from test_auroc_scores to test_f1_scores

# Perform cross-validation
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
for train_index, test_index in kf.split(df):
    # Reset the indices to ensure they are continuous
    train = df.iloc[train_index, :].reset_index(drop=True).copy()
    test = df.iloc[test_index, :].reset_index(drop=True).copy()

    response = 'y'  # the target is y
    predictors = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate',
       'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed',
       'day_of_week_encoded', 'education_encoded', 'month_encoded',
       'marital_divorced', 'marital_married', 'marital_single',
       'marital_unknown', 'job_admin.', 'job_blue-collar', 'job_entrepreneur',
       'job_housemaid', 'job_management', 'job_retired', 'job_self-employed',
       'job_services', 'job_student', 'job_technician', 'job_unemployed',
       'job_unknown', 'default_no', 'default_unknown', 'default_yes',
       'housing_no', 'housing_unknown', 'housing_yes', 'loan_no',
       'loan_unknown', 'loan_yes', 'contact_cellular', 'contact_telephone',
       'poutcome_failure', 'poutcome_nonexistent', 'poutcome_success']  # Replace with your predictor columns

    scaler = StandardScaler()

    X_train = scaler.fit_transform(train[predictors])
    X_test = scaler.transform(test[predictors])

    y_train = train[response]
    y_test = test[response]

    # Oversample the minority class in the training set
    oversampler = RandomOverSampler(random_state=42)
    X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train, y_train)

    # Define the logistic regression model
    logistic_reg = LogisticRegression()

    # Define the hyperparameter space for Bayesian optimization
    param_space = {
        'penalty': ['none', 'l2']
    }

    # Use Bayesian optimization to find the optimal hyperparameters
    opt = BayesSearchCV(logistic_reg, param_space, n_iter=50, cv=5)
    opt.fit(X_train_resampled, y_train_resampled)

    # Get the best hyperparameters
    best_params = opt.best_params_

    # Train the logistic regression model with the best hyperparameters
    logistic_reg2 = LogisticRegression(**best_params)
    logistic_reg2.fit(X_train_resampled, y_train_resampled)

    # Generate predictions for the training and test sets
    y_train_pred = logistic_reg2.predict(X_train)
    y_test_pred = logistic_reg2.predict(X_test)

    # Compute the F1 scores for training and test sets
    train_f1 = f1_score(y_train, y_train_pred)  # Use F1 score instead of AUROC
    test_f1 = f1_score(y_test, y_test_pred)

    train_f1_scores.append(train_f1)
    test_f1_scores.append(test_f1)

# Compute the mean of the F1 scores across all folds
mean_train_f1 = np.mean(train_f1_scores)
mean_test_f1 = np.mean(test_f1_scores)

# Print the best hyperparameters and the mean F1 scores
print("Best Hyperparameters:")
print(best_params)
print("\nTraining F1 Score:")
print(mean_train_f1)
print("\nTest F1 Score:")
print(mean_test_f1)
```
